```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataframe= pd.read_csv("Nike_Sales_Uncleaned.csv")
dataframe
```

```python
# DATA CLEANING AND PREPARING
```

```python
print(dataframe.isnull().sum())
```

```python
df=dataframe["Size"].dtype
df
```

```python
dataframe["Size"].fillna(method='ffill',inplace=True)
print(dataframe["Size"])
```

```python
print(dataframe.isnull().sum())
```

```python
df1=dataframe["Units_Sold"].dtype
df1
```

```python
dataframe['Units_Sold'].fillna(dataframe['Units_Sold'].mean(),inplace=True)
dataframe['MRP'].fillna(dataframe['MRP'].mean(),inplace=True)
print(dataframe['Units_Sold'])
print(dataframe['MRP'])
```

```python
print(dataframe.isnull().sum())
```

```python
dataframe["Discount_Applied"].fillna(0,inplace=True)
dataframe["Order_Date"].fillna(method='bfill',inplace=True)
```

```python
dataframe['Units_Sold'] = dataframe['Units_Sold'].round(2)
print(dataframe['Units_Sold'])
```

```python
print(dataframe.isnull().sum())
```

```python
#exploration
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.countplot(x=dataframe['Region'])
plt.xlabel("Different Region")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
grouped_data = dataframe.groupby('Region')['Profit'].sum()
result = pd.DataFrame({'Profit': grouped_data})
plt.plot(result, c='green', marker='o')
plt.xlabel('Different region', c='black', size=20)
plt.ylabel('Profit earned', c='black', size=20)
```

```python
max_Profit = dataframe['Profit'].max()
Region_with_max_profit = dataframe.loc[dataframe['Profit'] ==max_Profit, 'Product_Name']
print('Region with maximum Profit:')
print(Region_with_max_profit)
```

```python
sns.countplot(x=dataframe['Sales_Channel'])
```

```python
plt.hist(dataframe['Revenue'],bins=5)
plt.title('Revenue Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
couple_data=dataframe['Units_Sold']
sns.countplot(x=couple_data)
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'Gender_Category', y = 'Units_Sold', data = dataframe)
```

```python
pivot_table = dataframe.pivot_table(
    index='Region',
    columns='Sales_Channel',
    aggfunc='size',
    fill_value=0
)

sns.heatmap(pivot_table, annot=True, cmap='RdBu', fmt='d')
plt.title('Sales Distribution by Region and Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Region')
plt.show()
```
